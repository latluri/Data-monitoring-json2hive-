```python
final=df_today2.merge(df_yesterday2,indicator = True, how='outer',on=['TABLE_NAME','COLUMN_NAME'],suffixes=('_new','_old'))
final.rename(columns={'_merge':'Change'}, inplace=True)
```

```python
final["Change"]=final.Change.replace("left_only","added").replace("right_only","removed").replace("both","no change")
final.loc[(final['DATA_TYPE_old'] != final['DATA_TYPE_new']) & (final.Change =="no change"),"Change"]="dtype_change"
```

```python
dtype_changes=final.loc[(final['DATA_TYPE_old'] != final['DATA_TYPE_new']) & (final.Change =="dtype_change"),["TABLE_NAME","COLUMN_NAME","DATA_TYPE_old","DATA_TYPE_new","Change"] ]

add_removed=final.loc[ (final.Change =="added") | (final.Change =="removed"),["TABLE_NAME","COLUMN_NAME","DATA_TYPE_old","DATA_TYPE_new","Change"] ]
```

```python
result=pd.concat([dtype_changes,add_removed],ignore_index=True)
result["priority"]=result["TABLE_NAME"].apply(lambda r: priority(r))

if len(result) >0:
    result.to_excel("DESPPROD_schema_changesin24hrs"+today+".xlsx",index=False)
    commands.getstatusoutput("echo -e \"Schema changes in DES PPROD from yesterday are attached\"| mailx -s 'Changes in information schema -PPROD' -a \"DESPPROD_schema_changesin24hrs"+today+".xlsx\" latluri1@ford.com  ")
```
